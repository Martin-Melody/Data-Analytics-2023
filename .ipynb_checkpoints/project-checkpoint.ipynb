```python
# Importing necessary libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Setting display options for better readability
pd.set_option('display.width', 500)
pd.set_option('display.max_columns', 100)
```

## Dataset Description

This dataset compiles match data from the Abu Dhabi Combat Club (ADCC) Brazilian Jiu-Jitsu competitions, spanning from 1998 to 2022. It includes records from 1,028 matches, encapsulated in 14 distinct columns. Each column provides detailed insights into various aspects of the matches.

Key elements of the dataset are as follows:
- **Match ID**: A numerical identifier assigned uniquely to each match.
- **Winner and Loser Information**: This includes IDs and names of the participants, along with the points scored by the winners and losers.
- **Match Details**: Types of victory (e.g., decision, submission, points), specific submission techniques used (when applicable), and records of any advantages or penalties incurred.
- **Weight Class and Gender**: Classification of matches based on the competitors' weight class and gender.
- **Competition Stage and Year**: Information pertaining to the specific stage of the competition (e.g., final, semi-final) and the year of the match.

The dataset contains a blend of numerical and categorical data. Numerical data mainly includes match IDs and scoring points, while categorical data encompasses participant names, win types, submission techniques, etc. Notably, certain columns like 'submission' and 'adv_pen' (advantages/penalties) display missing values, which may reflect particular characteristics of the ADCC matches.

### Analysis of Most Popular Submissions

Here we analyze the types of submissions that occurred most frequently in the competitions. This gives an insight into the most effective or preferred submission techniques in Jiu-Jitsu.

```python
# Counting the frequency of each submission type
submission_counts = adcc_df['submission'].value_counts()

# Removing 

# Plotting the submission counts
plt.figure(figsize=(10, 6))
submission_counts.plot(kind='bar', color='skyblue')
plt.title("Most Popular Submissions")
plt.xlabel("Submission Type")
plt.ylabel("Number of Occurrences")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Get the top 10 submissions
top_10_submissions = submission_counts.head(10)

# Plotting the submission counts
plt.figure(figsize=(10, 6))
top_10_submissions.plot(kind='bar', color='skyblue')
plt.title("Top 10 Most Popular Submissions")
plt.xlabel("Submission Type")
plt.ylabel("Number of Occurrences")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### Identifying Best Scoring Athletes Each Year

In this section, we identify the athletes who scored the highest points each year. This highlights the top performers and their consistency over the years.

```python
# Filtering and grouping data to find the best athletes each year
filtered_data = adcc_df[adcc_df['year'].between(adcc_df['year'].min(), adcc_df['year'].max())]
best_athletes_each_year = filtered_data.loc[filtered_data.groupby('year')['winner_points'].idxmax()]

# Plotting the best scoring athletes
plt.figure(figsize=(12, 8))
bars = plt.bar(best_athletes_each_year['year'].astype(str), 
               best_athletes_each_year['winner_points'], 
               color='skyblue', alpha=0.7)
for bar, row in zip(bars, best_athletes_each_year.iterrows()):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5, row[1]['winner_name'],
             ha='center', va='bottom', rotation=90, fontsize=9)
plt.title(f"Best Scoring Athletes Each Year ({adcc_df['year'].min()} to {adcc_df['year'].max()})")
plt.xlabel("Year")
plt.ylabel("Points")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Statistical Analysis of Match Points

We now perform a statistical analysis on the points scored in matches. This includes calculating descriptive statistics for points scored by winners and losers in matches decided by points.

```python
# Filtering for matches decided by points and calculating statistics
points_matches = adcc_df[adcc_df['win_type'] == 'POINTS']
winner_points_stats = points_matches['winner_points'].describe()
loser_points_stats = points_matches['loser_points'].describe()

# Displaying the statistics
print("Winner Points Statistics:")
print(winner_points_stats)
print("\nLoser Points Statistics:")
print(loser_points_stats)
```
